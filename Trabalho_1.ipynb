**1.** Escreva um código em Python para calcular e exibir:

**a)** A soma dos números de 1 a 100.

```python
soma = 0
for i in range(1, 101):
    soma += i

print("A soma de 1 até 100 é:", soma)
```

**b)**  O perímetro de um círculo de raio 5, utilizando a fórmula \(P = 2\pi r\).

```python
import math

# Definição do raio
raio = 5

# Cálculo do perímetro
perimetro = 2 * math.pi * raio

# Exibição do resultado
print(f"O perímetro do círculo de raio {raio} é aproximadamente {perimetro:.2f}")
```

**c)** O fatorial de 6 sem utilizar bibliotecas externas.

```python
# Inicializando o fatorial
fatorial = 1

# Calculando o fatorial manualmente
for i in range(1, 7):  # 6! = 6 * 5 * 4 * 3 * 2 * 1
    fatorial *= i

# Exibindo o resultado
print("O fatorial de 6 é:", fatorial)
```

**d)** A área de um triângulo, utilizando uma função.

```python
def calcular_area_triangulo(base, altura):
    """Calcula a área de um triângulo dada a base e a altura."""
    return (base * altura) / 2

# Exemplo de uso
base = float(input("Digite a base do triângulo: "))
altura = float(input("Digite a altura do triângulo: "))

area = calcular_area_triangulo(base, altura)
print(f"A área do triângulo é: {area:.2f}")
```

**3.** Operações Matemáticas e Tratamento de Erros em Python:

**a)** Crie um programa que solicite dois números ao usuário e realize as quatro operações básicas.

```python
# Solicita dois números ao usuário
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Realiza as operações
soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2
divisao = num1 / num2

# Exibe os resultados
print("\nResultados das operações:")
print(f"Soma: {soma}")
print(f"Subtração: {subtracao}")
print(f"Multiplicação: {multiplicacao}")
print(f"Divisão: {divisao}")
```

**b)** Modifique o código para evitar erro de divisão por zero, exibindo uma mensagem apropriada.

```python
# Solicita dois números ao usuário
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Realiza as operações
soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2

# Evita divisão por zero
if num2 != 0:
    divisao = num1 / num2
else:
    divisao = "Indefinida (não é possível dividir por zero)"

# Exibe os resultados
print("\nResultados das operações:")
print(f"Soma: {soma}")
print(f"Subtração: {subtracao}")
print(f"Multiplicação: {multiplicacao}")
print(f"Divisão: {divisao}")
```

**4.** Validação de Entrada e Cálculos em Python:

**a)** Crie um programa que solicite um número ao usuário e valide se ele é um inteiro positivo.

```python
# Solicita um número ao usuário
num = input("Digite um número: ")

try:
    num = int(num)  # Tenta converter para inteiro
    if num > 0:
        print(f"O número {num} é um inteiro positivo.")
    else:
        print("O número não é positivo.")
except ValueError:
    print("Entrada inválida! Por favor, digite um número inteiro positivo.")
```

**b)** Se o valor inserido for válido, calcule e exiba o quadrado desse número. Se o valor inserido não for um inteiro positivo, exiba uma mensagem informando que a entrada é inválida.

```python
# Solicita um número ao usuário
num = input("Digite um número inteiro positivo: ")

try:
    num = int(num)  # Tenta converter para inteiro
    if num > 0:
        print(f"O quadrado de {num} é {num ** 2}.")
    else:
        print("Entrada inválida! O número deve ser um inteiro positivo.")
except ValueError:
    print("Entrada inválida! Por favor, digite um número inteiro positivo.")
```
