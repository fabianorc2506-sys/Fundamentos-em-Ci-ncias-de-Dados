```python
import pandas as pd
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv"
df = pd.read_csv(url)
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Dados das flores (comprimento, largura, espécie)
dados = np.array([
    [1.0, 0.2, "Setosa"],
    [1.5, 0.3, "Setosa"],
    [4.5, 1.5, "Versicolor"],
    [5.0, 1.7, "Versicolor"],
    [6.0, 2.0, "Virginica"]
])

# Divisão dos dados
X_train = dados[:, :2].astype(float)  # Comprimento e largura
y_train = dados[:, 2]  # Espécie

# Ponto a ser classificado
ponto_novo = np.array([[4.8, 1.6]])

# Aplicação do k-NN com k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
classe_predita = knn.predict(ponto_novo)

# Criar o gráfico
plt.figure(figsize=(8, 6))

# Plotar pontos reais
cores = {"Setosa": "blue", "Versicolor": "green", "Virginica": "purple"}
for especie in np.unique(y_train):
    subset = X_train[y_train == especie]
    plt.scatter(subset[:, 0], subset[:, 1], color=cores[especie], label=especie, marker="x", s=100)

# Plotar predição
plt.scatter(ponto_novo[:, 0], ponto_novo[:, 1], color="red", label=f"Predição ({classe_predita[0]})", marker="o", s=150)

# Configuração do gráfico
plt.title("Classificação k-NN (k=3) para a Flor")
plt.xlabel("Comprimento da pétala (cm)")
plt.ylabel("Largura da pétala (cm)")
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```
