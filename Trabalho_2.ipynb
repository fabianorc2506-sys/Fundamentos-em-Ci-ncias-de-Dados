**1.** Faça um sistema de votação em Python que funcione da seguinte maneira:



*   O sistema deve solicitar repetidamente o nome do candidato até que uma entrada vazia (``) seja digitada. Considere os candidatos como sendo "Corinthians", "Palmeiras", "São Paulo" e "Santos". Ofereça ainda a opção de votar em branco.
*   Cada usuário deve utilizar seu CPF para validar seu voto.
*   Utilize um dicionário para armazenar os votos de cada candidato.
*   Conte o número de votos de cada candidato.
*   Ao final, imprima a lista de todos os candidatos e seus respectivos votos.
*   Imprima também o nome do vencedor da eleição.

```python
def validar_cpf(cpf, cpfs_registrados):
    """Verifica se o CPF contém exatamente 11 dígitos e não foi usado."""
    if len(cpf) != 11 or not cpf.isdigit():
        print("CPF inválido. Deve conter exatamente 11 dígitos numéricos.")
        return False
    if cpf in cpfs_registrados:
        print("Este CPF já foi utilizado para votar.")
        return False
    return True

def main():
    candidatos = {"Corinthians": 0, "Palmeiras": 0, "São Paulo": 0, "Santos": 0, "Branco": 0}
    cpfs_registrados = set()

    while True:
        cpf = input("Digite seu CPF (ou pressione Enter para finalizar a votação): ")
        if cpf == "":
            break

        if not validar_cpf(cpf, cpfs_registrados):
            continue

        print("Candidatos: Corinthians, Palmeiras, São Paulo, Santos, Branco")
        voto = input("Digite o nome do candidato: ").strip()

        if voto in candidatos:
            cpf_confirmacao = input("Digite novamente seu CPF para confirmar o voto: ")
            if cpf_confirmacao == cpf:
                candidatos[voto] += 1
                cpfs_registrados.add(cpf)
                print("Voto computado com sucesso!\n")
            else:
                print("Os CPFs não correspondem. Tente novamente.\n")
        else:
            print("Candidato inválido. Tente novamente.\n")

    print("\nResultados da eleição:")
    for candidato, votos in candidatos.items():
        print(f"{candidato}: {votos} votos")

    vencedor = max(candidatos, key=candidatos.get)
    print(f"\nO vencedor da eleição é: {vencedor}")

if __name__ == "__main__":
    main()
```

**3.** Escreva um código em Python que, dadas duas matrizes arbitrárias $A$ e $B$, calcule o produto delas. Não esqueça de verificar se o produto entre as matrizes está bem definido. Faça o produto manualmente, isto é, sem usar comandos da biblioteca numpy.

```python
def verificar_dimensoes(A, B):
    """Verifica se o produto de A e B está bem definido"""
    return len(A[0]) == len(B)

def multiplicar_matrizes(A, B):
    """Realiza a multiplicação manual de duas matrizes"""
    if not verificar_dimensoes(A, B):
        print("Erro: As matrizes não podem ser multiplicadas devido às dimensões incompatíveis.")
        return None

    linhas_A = len(A)
    colunas_A = len(A[0])
    colunas_B = len(B[0])

    # Inicializando matriz resultado com zeros
    resultado = [[0] * colunas_B for _ in range(linhas_A)]

    # Multiplicação das matrizes
    for i in range(linhas_A):
        for j in range(colunas_B):
            for k in range(colunas_A):  # ou len(B), já que colunas_A == linhas_B
                resultado[i][j] += A[i][k] * B[k][j]

    return resultado

# Exemplo de entrada
A = [[1, 2], [3, 4]]
B = [[-5, 6], [7, 8]]

# Calculando o produto das matrizes
resultado = multiplicar_matrizes(A, B)

# Exibindo o resultado
if resultado:
    print("\nMatriz produto:")
    for linha in resultado:
        print(linha)
```
